```python
import time
from tqdm import tqdm
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import warnings
warnings.filterwarnings('ignore')


chrome_options = Options()
chrome_options.add_argument('--log-level=3')
chrome_options.add_argument("--no-sandbox")
chrome_options.add_experimental_option("detach", True)


chrome_options.add_argument("--profile-directory=Default")
chrome_options.add_argument("--user-data-dir=C:\\Users\\Bill\\AppData\\Local\\Google\\Chrome\\User Data")

chrome_options.headless = False
#chrome_options.add_experimental_option("debuggerAddress", "127.0.0.1:9222")
#wd_Chrome = webdriver.Chrome("/usr/bin/chromedriver", ChromeDriverManager().install(),options=chrome_options,)

wd_Chrome = webdriver.Chrome(options=chrome_options)


```

```python
# Com o WebDrive a gente consegue a pedir a p√°gina (URL)
wd_Chrome.get("https://www.tiktok.com/creator-center/analytics/content/trending")
time.sleep(15)
```

```python
# Encontrar todos os elementos com o mesmo seletor
elementos = 0
df = pd.DataFrame(columns=['T√≠tulo', 'Tempo Medio de Visualiza√ß√£o', 'Seguidores', 'Porc_video_completo'])

while elementos < 9:

    # Execute as a√ß√µes desejadas em cada elemento
    elemento =  wd_Chrome.find_elements(By.CLASS_NAME,'css-1et5rn1')[elementos]
    elemento.click()
    time.sleep(5)

    titulo = wd_Chrome.find_element(By.CLASS_NAME,'css-1f3riye').text
    tempo_medio = wd_Chrome.find_elements(By.CLASS_NAME,'css-1pot8pv')[1].text
    seguidores = wd_Chrome.find_elements(By.CLASS_NAME,'css-1pot8pv')[2].text
    Porc_video_completo = wd_Chrome.find_elements(By.CLASS_NAME,'css-1pot8pv')[3].text

    titulo, hastags = titulo.split('#', 1)

    linha = {'T√≠tulo': titulo, 'Tempo Medio de Visualiza√ß√£o': tempo_medio, 'Seguidores':seguidores,'Porc_video_completo': Porc_video_completo }
    df = df.append(linha, ignore_index=True)
    # print(titulo)
    # print(tempo_medio)
    # print(seguidores)
    # print(Porc_video_completo)


    # Navegar de volta para a p√°gina anterior
    wd_Chrome.back()
    time.sleep(2)
    # wd_Chrome.back()
    # time.sleep(2)

    elementos = elementos+1

```

```python
df2 = pd.read_excel(f'Outubro\\Trending Videos final.xlsx')
```

```python


resultado = pd.merge(df, df2, on='T√≠tulo', how='inner')

# Usar uma fun√ß√£o lambda para remover o sufixo "K" ou "M" e converter para int
resultado['Seguidores'] = resultado['Seguidores'].apply(lambda texto: int(float(texto[:-1]) * 1000) if texto.endswith('K') else int(float(texto)))

```

```python
#resultado['Seguidores'] = resultado['Seguidores'].astype(int)
resultado['Porc_seguidores_View'] = resultado.apply(lambda row: round((row['Seguidores'] / row['Total views']) * 100, 2), axis=1)

resultado.to_excel(f'Outubro\\Trending Videos final2.xlsx', index=False)
resultado
```
